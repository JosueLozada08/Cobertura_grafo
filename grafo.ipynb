```python
def flujo_control(a, b, c):
    resultado = ""

    # bloque de decisiones 
    if a > 0:
        if b > 0:
            resultado += "R1-"
        else:
            resultado += "R2-"
    else:
        if c > 0:
            resultado += "R3-"
        else:
            resultado += "R4-"

    # Segunda decisión (
    if a + b + c > 5:
        resultado += "R5-"
    else:
        resultado += "R6-"

    # Última decisión 
    if a == b:
        resultado += "R7"
    else:
        resultado += "R8"

    return resultado



def pruebas():
    print("Caso 1: a=1, b=1, c=0")      # R1–R5–R8
    print("Salida:", flujo_control(1, 1, 0))
    print()

    print("Caso 2: a=1, b=-1, c=2")    # R2–R6–R8
    print("Salida:", flujo_control(1, -1, 2))
    print()

    print("Caso 3: a=-1, b=0, c=2")    # R3–R6–R8
    print("Salida:", flujo_control(-1, 0, 2))
    print()

    print("Caso 4: a=-1, b=0, c=-1")   # R4–R6–R8
    print("Salida:", flujo_control(-1, 0, -1))
    print()

    print("Caso 5: a=2, b=2, c=2")     # R1–R5–R7
    print("Salida:", flujo_control(2, 2, 2))
    print()

# Ejecutar pruebas
if __name__ == "__main__":
    pruebas()
```

```python
import matplotlib.pyplot as plt
import networkx as nx


G = nx.DiGraph()


nodos = range(1, 14)
G.add_nodes_from(nodos)


edges = [
    (1, 2),
    (2, 3), (2, 4),
    (3, 5), (4, 5),
    (5, 6),
    (6, 7), (6, 8),
    (7, 9), (8, 9),
    (9, 10),
    (10, 11), (10, 12),
    (11, 13), (12, 13)
]
G.add_edges_from(edges)

pos = {
    1: (0, 12),
    2: (0, 11),
    3: (-1, 10),
    4: (1, 10),
    5: (0, 9),
    6: (0, 8),
    7: (-1, 7),
    8: (1, 7),
    9: (0, 6),
    10: (0, 5),
    11: (-0.5, 4),
    12: (0.5, 4),
    13: (0, 3)
}

# Dibujar nodos y aristas
plt.figure(figsize=(8, 10))
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=800)
nx.draw_networkx_edges(G, pos, arrows=True, arrowstyle='->', edge_color='black')
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

# Mostrar el grafo
plt.title("Grafo de flujo de control")
plt.axis('off')
plt.tight_layout()
plt.show()
```
